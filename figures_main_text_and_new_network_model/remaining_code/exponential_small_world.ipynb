```python
G = nx.empty_graph(N)
mean = 20
N = 20000
degree_seq = [ int(k) for k in expon.rvs(scale=mean,size=N)]
stubs = list(degree_seq)
print(sum(stubs))
if sum(stubs) % 2 == 1:
    stubs[np.random.randint(0,N-1)] += 1

order = np.argsort(stubs)[::-1]
order = np.random.permutation(N)


edges = []
for i in order:
    d = 1
    up = True
    while stubs[i] > 0:
        if up:
            j = (i+d) % N
        else:
            j = (i-d) % N
            d += 1
            
        if stubs[j] > 0:
            edges.append((i,j))
            stubs[i] -= 1
            stubs[j] -= 1
        up = not up
G.add_edges_from(edges)
#print(longest_distance(G))
```

```python
def longest_distance(G):
    max_l = -1
    for u, v in G.edges():
        d = np.abs(u-v)
        d = min(d,N-d)
        max_l = max(max_l,d)
    return max_l
def get_distances(G):
    rs = []
    for u, v in G.edges():
        d = np.abs(u-v)
        d = min(d,N-d)
        rs.append(d)
    return rs
```
