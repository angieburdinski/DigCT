```python
import netwulf as nw
import networkx as nx
from scipy.stats import binom, expon
import numpy as np
```

```python
N = 21
mean = 4
k = np.arange(N)
p = 1/mean * np.exp(k/mean)
deg_seq = [ int(expon.rvs(scale=mean)) for n in range(N) ]
G = nx.configuration_model(deg_seq)
```

```python
result = nw.visualize(nx.Graph(G))
```
